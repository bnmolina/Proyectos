```python
# Implementación de un perceptrón simple

class Perceptron:
    
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    # cálculo del puntaje z
    def calcular_z(self, inputs):
        z = 0
        for w, x in zip(self.weights, inputs):
            z += w * x
        z += self.bias
        return z

    # función de activación (clasificación)
    def predecir(self, inputs):
        z = self.calcular_z(inputs)
        if z >= 0:
            return 1, z
        else:
            return 0, z


# Definir pesos y bias
pesos = [0.6, 0.8]
bias = -0.7

modelo = Perceptron(pesos, bias)
```

```python
# Casos de prueba
pruebas = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1],
    [0.5, 0.5],
    [0.2, 0.9]
]

print("Entradas | Puntaje z | Salida")

for entrada in pruebas:
    salida, z = modelo.predecir(entrada)
    print(entrada, "|", round(z,2), "|", salida)
```

Como conclusiones podemos decir que

1. El perceptrón clasifica los datos utilizando un puntaje lineal (z) .

2. Los pesos controlan la influencia de cada característica en la decisión.

3. El sesgo desplaza la frontera de decisión , afectando qué tan fácil es obtener salida 1.

4. Pequeños cambios en pesos o sesgos pueden cambiar completamente la clasificación de ciertos casos
