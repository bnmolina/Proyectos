```python
import numpy as np
import pandas as pd

def perception(x, w, b):
    
    #calculamos el puntaje
    z = np.dot(x, w) + b
    
    #función de activación
    if z >=0:
        salida = 1 
    else:
        salida = 0
        
    return z, salida  
#W pesos
w = np.array([1, 1])

#B bias
b = -1

pruebas = [
    [0, 0], [0, 1], [1, 0], [1, 1]
]

re = []

for p in pruebas:
    
    x = np.array(p)
    z, salida = perception(x, w, b)
    
    re.append([p[0], p[1], z, salida])
    
    print("Entrada: ", p, "Z: ", z, "Salida: ", salida)
    tabla = pd.DataFrame(re, columns = ["X1", "X2", "Z", "SALIDA"])
    tabla
    
tabla    
```

Como conclusiones podemos decir que

1. El perceptrón clasifica los datos utilizando un puntaje lineal (z) .

2. Los pesos controlan la influencia de cada característica en la decisión.

3. El sesgo desplaza la frontera de decisión , afectando qué tan fácil es obtener salida 1.

4. Pequeños cambios en pesos o sesgos pueden cambiar completamente la clasificación de ciertos casos
