```python
# Importar librerías
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Cargar dataset simple (MNIST)
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalizar datos
x_train = x_train / 255.0
x_test = x_test / 255.0

# Crear modelo sencillo
def crear_modelo():
    model = keras.Sequential([
        keras.layers.Flatten(input_shape=(28,28)),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])
    return model

# Configuración 1: learning rate alto
modelo1 = crear_modelo()
modelo1.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.01),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Entrenando modelo con learning rate = 0.01")
hist1 = modelo1.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))


# Configuración 2: learning rate bajo
modelo2 = crear_modelo()
modelo2.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nEntrenando modelo con learning rate = 0.001")
hist2 = modelo2.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))


# Graficar resultados
plt.plot(hist1.history['loss'], label='LR=0.01')
plt.plot(hist2.history['loss'], label='LR=0.001')
plt.title('Comparación de pérdida (loss)')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# semana 5 - Hiperparámetros

Comparar el efecto de la tasa de aprendizaje en el entrenamiento de una red neuronal.

Se comparo
- Learning rate = 0.01
- Learning rate = 0.001

El resultado principal
El learning rate más bajo (0.001) mostró mayor estabilidad durante el entrenamiento.
